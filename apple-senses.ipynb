```python
import matplotlib.pyplot as plt
import networkx as nx
%matplotlib inline
```

```python
G = nx.DiGraph()
G.add_nodes_from([
    'entity.n.01',
    'physical_entity.n.01',
    'abstraction.n.06',
    'natural_object.n.01',
    'living_thing.n.01',
    'social_group.n.01',
    'artifact.n.01',
    'fruit.n.01',
    'tree.n.01',
    'company.n.01',
    'machine.n.01',
    'apple.n.01',
    'apple.n.02',
    'apple.n.03',
    'apple.n.04'
])

G.add_edges_from([
    ('entity.n.01', 'physical_entity.n.01'),
    ('entity.n.01', 'abstraction.n.06'),
    ('physical_entity.n.01', 'natural_object.n.01'),
    ('physical_entity.n.01', 'living_thing.n.01'),
    ('abstraction.n.06', 'social_group.n.01'),
    ('abstraction.n.06', 'artifact.n.01'),
    ('natural_object.n.01', 'fruit.n.01'),
    ('living_thing.n.01', 'tree.n.01'),
    ('social_group.n.01', 'company.n.01'),
    ('artifact.n.01', 'machine.n.01'),
    ('fruit.n.01', 'apple.n.01'),
    ('tree.n.01', 'apple.n.02'),
    ('company.n.01', 'apple.n.03'),
    ('machine.n.01', 'apple.n.04')
])

nx.draw(G, with_labels=True, font_size=6)
plt.show()
```

```python
# Write graph to file
nx.write_gml(G, 'data/graphs/apple-senses.gml')
```
